```python
rating = pd.DataFrame(df.groupby('title').count()['rating'])
```

```python
rating['number of rating'] = pd.DataFrame(df.groupby('title').count()['rating'])
```

```python
rating.sort_values(by='rating',ascending=False)
```

```python
moviemat = df.pivot_table(index='user',columns='title',values='rating')
```

```python
moviemat
```

```python
def predict_movies(movie_name):
    movie_user_rating = moviemat[movie_name]
    similar_to_movie = moviemat.corrwith(movie_user_rating)
    corr_movie = pd.DataFrame(similar_to_movie,columns=['correlation'])
    corr_movie.dropna(inplace=True)
    corr_movie = corr_movie.join(rating['number of rating'])
    predictions = corr_movie[corr_movie['number of rating']>100].sort_values('correlation',ascending = False)
    return predictions
```

```python
predictions = predict_movies(input())
```
